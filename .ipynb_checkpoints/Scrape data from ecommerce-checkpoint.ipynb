```python
import requests
from bs4 import BeautifulSoup
```

```python

def scrape_amazon_discounts():
    url = 'https://www.amazon.com/s?k=discounts'

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    products = []

    for item in soup.find_all('div', {'data-cy': 's-search-result'}):
        name = item.find('span', {'class': 'a-size-mini a-spacing-none a-color-base s-line-clamp-4'}).text
        price = item.find('span', {'class': 'a-offscreen'}).text
        link = 'https://www.amazon.com' + item.find('a', {'class': 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})['href']

        # Check if the product is discounted
        if item.find('span', {'class': 'a-text-price'}):
            original_price = item.find('span', {'class': 'a-price'}).find('span', {'class': 'a-offscreen'}).text
            discount = item.find('span', {'class': 'a-text-price'}).text
            products.append({'Name': name, 'Price': price, 'Original Price': original_price, 'Discount': discount, 'Link': link})

    return products

amazon_discounts = scrape_amazon_discounts()

for product in amazon_discounts:
    print(product)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_jumia_discounts():
    url = 'https://www.jumia.co.ke/catalog/?q=discounts'

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    products = []

    for item in soup.find_all('article', class_='prd _fb col c-prd'):
        name = item.find('h3', class_='name').text.strip()
        price = item.find('div', class_='prc').text.strip()
        link = item.find('a', class_='core')['href']

        # Check if the product is discounted
        discount_tag = item.find('div', class_='dsct')
        if discount_tag is not None:
            original_price = item.find('div', class_='old').text.strip()
            discount = discount_tag.text.strip()
            products.append({'Name': name, 'Price': price, 'Original Price': original_price, 'Discount': discount, 'Link': link})
        else:
            products.append({'Name': name, 'Price': price, 'Link': link})

    return products

jumia_discounts = scrape_jumia_discounts()

for product in jumia_discounts:
    print(product)
```
