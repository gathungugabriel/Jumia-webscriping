```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def scrape_jumia_discounts(pages=50):
    base_url = 'https://www.jumia.co.ke/all-products/?page={}'
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

    products = []

    for page_num in range(1, pages+1):
        url = base_url.format(page_num)
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')

        for item in soup.find_all('article', class_='prd _fb col c-prd'):
            name = item.find('h3', class_='name').text.strip()
            price = item.find('div', class_='prc').text.strip()
            link = item.find('a', class_='core')['href']
            old_price_tag = item.find('div', class_='old')
            bonus_tag = item.find('div', class_='bdg _dsct _sm')

            # Check if the product has a bonus (based on the presence of the "old" class)
            if old_price_tag:
                old_price = old_price_tag.text.strip()
                bonus = bonus_tag.text.strip() if bonus_tag else None
                products.append({'Name': name, 'Price': price, 'Old Price': old_price, 'Bonus': bonus, 'Link': link})

    return products

# Scrape Jumia discounts with bonuses from all 50 pages
jumia_discounts_with_bonus = scrape_jumia_discounts(pages=50)

# Create DataFrame
df = pd.DataFrame(jumia_discounts_with_bonus)

# Display DataFrame
print(df)
```

```python
import pandas as pd
import re

# Load the data
data = pd.read_csv('jumia_products.csv')  # Replace 'your_data.csv' with the actual file path

# Function to extract numeric values from strings using regular expressions
def extract_numeric(string):
    match = re.search(r'\d+', string)
    if match:
        return int(match.group())
    else:
        return None

# Remove 'KSh' prefix and split by '-'
data['Price'] = data['Price'].str.replace('KSh', '').str.strip()
data['Old Price'] = data['Old Price'].str.replace('KSh', '').str.strip()

price_split = data['Price'].str.split('-', expand=True)
old_price_split = data['Old Price'].str.split('-', expand=True)

# Extract numeric values from the split strings
data['Price Min'] = price_split[0].apply(extract_numeric)
data['Price Max'] = price_split[1].apply(lambda x: extract_numeric(x) if x else None)
data['Old Price Min'] = old_price_split[0].apply(extract_numeric)
data['Old Price Max'] = old_price_split[1].apply(lambda x: extract_numeric(x) if x else None)

# Fill NaN values in the new columns with the original values
data['Price Min'].fillna(data['Price'], inplace=True)
data['Price Max'].fillna(data['Price'], inplace=True)
data['Old Price Min'].fillna(data['Old Price'], inplace=True)
data['Old Price Max'].fillna(data['Old Price'], inplace=True)

# Drop the original Price and Old Price columns
data.drop(columns=['Price', 'Old Price'], inplace=True)

# Display a sample of the data
data.sample(30)
```
