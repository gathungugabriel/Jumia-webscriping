```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re

def extract_numeric(string):
    match = re.search(r'\d+', string)
    if match:
        return int(match.group())
    else:
        return None

def scrape_jumia_discounts():
    url_template = 'https://www.jumia.co.ke/all-products/?page={}'
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    products = []

    for page_num in range(1, 51):  # Loop through all 50 pages
        url = url_template.format(page_num)
        print(f"Scraping page {page_num}: {url}")  # Add print statement
        response = requests.get(url, headers=headers)
        print(f"Response status code: {response.status_code}")  # Add print statement
        soup = BeautifulSoup(response.content, 'html.parser')

        for item in soup.find_all('article', class_='prd _fb col c-prd'):
            name = item.find('h3', class_='name').text.strip()
            price = item.find('div', class_='prc').text.strip()
            link = item.find('a', class_='core')['href']

            # Check if the product is discounted
            discount_tag = item.find('div', class_='bdg _dsct _sm')
            if discount_tag is not None:
                discount = discount_tag.text.strip()
                # Extract the old price tag if it exists
                old_price_tag = item.find('div', class_='old')
                old_price = old_price_tag.text.strip() if old_price_tag else None
                products.append({'Name': name, 'Price': price, 'Old Price': old_price, 'Discount': discount, 'Link': link})

    return products

# Call the function and print the results
result = scrape_jumia_discounts()
result
```

```python
import pandas as pd

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(result)

# Save the DataFrame to a CSV file
df.to_csv('jumia_products.csv', index=False)

print("CSV file created successfully.")
```

DATA CLEANING

```python
# Convert 'Price' and 'Old Price' columns to string type
data['Price'] = data['Price'].astype(str)
data['Old Price'] = data['Old Price'].astype(str)

# Remove 'KSh' prefix, commas, and spaces
data['Price'] = data['Price'].str.replace('KSh', '').str.replace(',', '').str.replace(' ', '')
data['Old Price'] = data['Old Price'].str.replace('KSh', '').str.replace(',', '').str.replace(' ', '')

# Define a function to convert to numeric with error handling
def convert_to_numeric(value):
    if '-' in value:  # If the value is a range
        range_values = value.split('-')
        try:
            lower_bound = float(range_values[0])
            upper_bound = float(range_values[1])
            return (lower_bound + upper_bound) / 2  # Take the average of the range
        except ValueError:
            print(f"Error converting value: {value}")
            return None
    else:
        try:
            return float(value)
        except ValueError:
            print(f"Error converting value: {value}")
            return None

# Convert 'Price' and 'Old Price' to numeric with error handling
data['Price'] = data['Price'].apply(convert_to_numeric)
data['Old Price'] = data['Old Price'].apply(convert_to_numeric)

# Display the DataFrame
data
```
